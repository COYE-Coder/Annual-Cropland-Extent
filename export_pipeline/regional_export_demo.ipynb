# Regional Export Pipeline for Landsat Time Series

This notebook demonstrates how to export regional Landsat time series data using the Earth Engine Python API. The pipeline:

1. Processes Landsat imagery across missions (5, 7, and 8)
2. Combines with RAP vegetation cover data
3. Creates spatially stratified export grids
4. Exports data as Earth Engine assets

## Setup and Configuration

```python
import ee
from src.export_utils import process_and_export_regional

# Initialize Earth Engine
ee.Initialize()

# Asset paths
MEXICO_COVER_ASSET = '////REDACTED////'
CONUS_COVER_ASSET = 'projects/rangeland-analysis-platform/vegetation-cover-v3'
OUTPUT_ASSET_PATH = 'projects/YOUR_EE_ASSET_FOLDER/'

# Define export geometry
GEOM_TO_EXPORT = ee.Geometry.Polygon([
    [[-100.06872998114088, 53.54840419627541],
     [-116.06482373114088, 53.96410259798528],
     [-116.06482373114088, 48.931452079494676],
     [-129.95154248114088, 47.28847269264896],
     [-114.48279248114088, 19.324002572629446],
     [-91.01599560614086, 25.177505795258153],
     [-90.84021435614086, 40.85736734731578],
     [-93.56482373114088, 48.35070265825225],
     [-98.83826123114088, 49.27669529994239]]])
```

## Create Export Grids

```python
# Process and export regional data
process_and_export_regional(
    geom_to_export=GEOM_TO_EXPORT,
    mexico_cover_asset=MEXICO_COVER_ASSET,
    conus_cover_asset=CONUS_COVER_ASSET,
    output_asset_path=OUTPUT_ASSET_PATH
)
```

## Notes

- The export process can take several hours to complete depending on the size of the region
- Each grid cell is exported as a separate Earth Engine asset
- Assets include both Landsat spectral indices and RAP vegetation cover data, which you may not have access to.
- Data is processed for late summer/fall (August-November) of each year
